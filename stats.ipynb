```python
df[numerical_cols].hist(figsize=(14, 12), bins=20, color='blue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
df['crash_date'] = pd.to_datetime(df['crash_date']) #convert crash_date to datetime format
```

```python
plt.figure(figsize = (12,5))
df['crash_date'].dt.to_period("M").value_counts().sort_index().plot()
plt.title("Monthly Traffic Accidents Trend")
plt.xlabel("Date")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show() #accident over time
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['prim_contributory_cause'],order = df['prim_contributory_cause'].value_counts().index[:10])
plt.title("Top 10 Causes of Accidents")
plt.xlabel("Count")
plt.ylabel("Cause")
plt.show() #most common cause of accidents
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['most_severe_injury'],order=df['most_severe_injury'].value_counts().index[:10])
plt.title("Accident Severity Distribution")
plt.xlabel("Count")
plt.ylabel("Severity")
plt.show() #severe accidents distribution
```

```python
plt.figure(figsize=(12,5))
sns.countplot(y=df['weather_condition'],order=df['weather_condition'].value_counts().index[:10])
plt.title("Accidents by Weather Condition")
plt.xlabel("Count")
plt.ylabel("Weather condition")
plt.show() #accidents with weather condition
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['crash_hour'],bins=24,kde=True)
plt.title("Accidents by Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.xticks(range(0,24))
plt.show() #accidents with hour of the day
```

```python
plt.figure(figsize =(12,5))
sns.countplot(x='roadway_surface_cond', hue='most_severe_injury',data=df,order=df['roadway_surface_cond'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Accident Severity by Weather Condition")
plt.xlabel("Weather Conditon")
plt.ylabel("Total Injuries")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x='crash_hour',y='injuries_total',data=df)
plt.title("Accident Severity by Hour of the Day")
plt.xlabel("Hour of Day")
plt.ylabel("Total Injuries")
plt.show()
```

```python
plt.figure(figsize=(12,5))

numeric_df = df.select_dtypes(include =['number'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='magma',fmt ='.2f',linewidth =0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
sns.pairplot(df, vars=['injuries_total', 'injuries_non_incapacitating', 'injuries_reported_not_evident'], hue='weather_condition', palette='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['injuries_total'], y=df['injuries_non_incapacitating'], hue=df['roadway_surface_cond'])
plt.title("Scatter Plot: injuries total vs injuries non incapacitating")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['injuries_total'], y=df['injuries_reported_not_evident'], hue=df['roadway_surface_cond'])
plt.title("Scatter Plot: injuries total vs injuries reported not evident")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['crash_hour'], y=df['injuries_total'], hue=df['roadway_surface_cond'])
plt.title("Scatter Plot: crash hour vs injuries_total")
plt.show()
```
